```python
import csv
```

```python
from chembl_webresource_client.new_client import new_client
```

```python
# This will be resulting structure mapping target ChEMBL IDs into 
# uniprot IDs
chem2uni = dict()
```

```python
# First, parse the csv file to extract target ChEMBL IDs:
with open('data/chembl_moa_target_chembl_id.csv', 'rb') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        chem2uni[row[0]] = set()
```

```python
chem2uni
```

```python
keys = list(chem2uni.keys())
```

```python
for key in keys:
    targets = new_client.target.filter(target_chembl_id__in=key).only(
            ['target_components'])
    uniprots=set()
    uniprots |= set(
            sum([[comp['accession'] for comp in 
                  t['target_components']] for t in targets],[]))
    chem2uni[key] = uniprots
```

```python
chem2uni
```

```python
len(chem2uni)
```

```python
# Finally write it to the output csv file
with open('data/chembl_moa_target_chem2uni.csv', 'wb') as csvfile:
    writer = csv.writer(csvfile)
    i=0
    for key, val in chem2uni.items():
        print(key, val)
        print(i)
        i = i+1
        val.discard(None)
        writer.writerow([key] + [str('; '.join(list(val)))])
```
